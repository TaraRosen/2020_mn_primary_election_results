```python
import sys
import requests
from bs4 import BeautifulSoup

import pandas as pd 

pd.set_option("display.max_rows", None, "display.max_columns", None)
```

```python
# import csv with precinct codes and county ids for CD5

cd5_url_codes = pd.read_csv('cd5.csv')
```

```python
cd5_pc_cid_codes = pd.read_csv('cd5_pc_cid.csv')
```

```python
# create list of CD5 precinct codes

cd5_pc = cd5_pc_cid_codes["precinct_code"].to_list()

# create list of CD5 county codes

cd5_cc = cd5_pc_cid_codes["county_code"].to_list()

print(cd5_pc[:6])
print(cd5_cc[:6])
```

```python
# create list of CD5  

cd5_cd = cd5_url_codes["cd"].to_list()

# create list of CD5 county

cd5_county = cd5_url_codes["county"].to_list()

# create list of CD5 precinct name

cd5_precinct = cd5_url_codes["precinct_name"].to_list()


print(cd5_cd[:6])
print(cd5_county[:6])
print(cd5_precinct[:6])
```

```python
# change electionid to 135

urls = []
county_id = [2, 2]
precinct_code = [1238977, 1238978]
cd5_county[:2]
cd5_cd[:2]
cd5_precinct[:2]
    
for i in range(len(county_id)):

        url = 'https://electionresults.sos.state.mn.us/results/Index?ErsElectionId=115&CountyId=' + str(county_id[i]) + '&DistrictId=&Scenario=Precincts&selectprecincts='+ str(precinct_code[i]) + '&show=Show+Selected+Precincts'
        urls.append(url)
        
        results = []
        vote_totals = []

        for u in urls:
            page = requests.get(u)
            soup_3 = BeautifulSoup(page.content, 'html.parser')

            rows = soup_3.find_all('tr')


            for row in rows:

                party_long = row.find(attrs={"class": "hidden-sm hidden-xs col-md-3 text-right"})
                party_short = row.find(attrs={"class": "hidden-lg hidden-md col-sm-1 col-xs-2 text-right"})
                candidate = row.find(attrs={"class": "col-md-3 col-sm-3 col-xs-3"})
                raw_votes = row.find(attrs={"class":"col-md-2 col-sm-2 col-xs-2 text-right"})
        #         for vote_total in row.find_all(attrs={"class":"col-md-2 col-sm-2 col-xs-2 text-right"}):
        #             vote_totals.append(vote_total)

                results.append([cd5_cd[i], cd5_county[i], cd5_precinct[i], party_long, party_short, candidate, raw_votes])

                results_df = pd.DataFrame(results, columns = ['CD', 'County', 'Precinct', 'Party_Long', 'Party_Short', 'Candidate', 'Raw_Votes'])
results_df
```
