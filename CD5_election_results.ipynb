```python
# import sys
# import requests
# from bs4 import BeautifulSoup

# import pandas as pd 

pd.set_option("display.max_rows", None, "display.max_columns", None)


def election_results(county_id, precinct_id):    
    
    # election results for all precincts

    test_urls = ['https://electionresults.sos.state.mn.us/results/Index?ErsElectionId=115&CountyId=' + str(county_id) + '&DistrictId=&Scenario=Precincts&selectprecincts='+ str(precinct_id) + '&show=Show+Selected+Precincts']

    results = []
    vote_totals = []

    for url in test_urls:
        page = requests.get(url)
        soup_3 = BeautifulSoup(page.content, 'html.parser')
    
        rows = soup_3.find_all('tr')

    
        for row in rows:
        
            party_long = row.find(attrs={"class": "hidden-sm hidden-xs col-md-3 text-right"})
            party_short = row.find(attrs={"class": "hidden-lg hidden-md col-sm-1 col-xs-2 text-right"})
            candidate = row.find(attrs={"class": "col-md-3 col-sm-3 col-xs-3"})
            raw_votes = row.find(attrs={"class":"col-md-2 col-sm-2 col-xs-2 text-right"})
    #         for vote_total in row.find_all(attrs={"class":"col-md-2 col-sm-2 col-xs-2 text-right"}):
    #             vote_totals.append(vote_total)
            
            results.append([party_long, party_short, candidate, raw_votes])
    
            results_df = pd.DataFrame(results, columns = ['Party_Long', 'Party_Short', 'Candidate', 'Raw_Votes'])

    return results_df
```
